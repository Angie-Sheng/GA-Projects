In this part, I would like to clean the data sets I obtained from web scraping. Because I use the same methdology for scrapying different jobs, so the structure for each job is same. I will use the data set containing data analyst jobs as an example to demonstate the cleaning process and do the same thing for other jobs.

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
data_analyst = pd.read_csv('./data_analyst_scraping.csv') #loading data set
```

```python
data_analyst.head() #a glimpse at the data set
```

Check the 'Job Description', that's where I store my output stright away after extract text from job descriptions

```python
data_analyst['Job Description'][0]
```

That seems not correct because some of the words stick to each other, e.g., 'expertiseWe'. Luckily, I keep the raw material in column "JD".

```python
data_analyst = data_analyst.drop(columns=['Job Description'])
```

```python
data_analyst['JD'][0] #check 'JD' column
```

```python
#try to feed the raw material into BeautifulSoup and check the result
r = data_analyst['JD'][0]
Soup = BeautifulSoup(r, 'html')
test_JD = Soup.get_text(separator=" ", strip=True)
```

```python
test_JD
```

This seems much better, except some annoying "\xa03". 

```python
#for replacement
import re

def replace(string, substitutions):

    substrings = sorted(substitutions, key=len, reverse=True)
    regex = re.compile('|'.join(map(re.escape, substrings)))
    return regex.sub(lambda match: substitutions[match.group(0)], string)
```

```python
string = test_JD
repla = {u'\xa0':'  ',u'\xe2\x80\x9d':'  ', u'\n':'  '}
replace(string,repla)
```

This is perfect! No sticking to each other and clean! I'm going to apply this to the rest.

```python
JD = []
for i in data_analyst['JD']:
    try:
        JD.append(replace(BeautifulSoup(i, 'html').get_text(separator=" ", strip=True),repla))
    except:
        JD.append('N/A')
        
data_analyst['Job Description'] = JD
```

```python
data_analyst.head()
```

```python
data_analyst.columns
```

I choose to drop the following columns: 'Unnamed: 0','Salary', 'Salary Type','JD'. There are a lot missing values in "Salary", so when I scraped the website, I add the filter "Salary Range". That's why we can drop "Salary" and "Salary Type" here because they cannot provide enough information.

```python
drop_columns = ['Unnamed: 0','Salary', 'Salary Type','JD','Title']
data_analyst = data_analyst.drop(columns=drop_columns)
```

```python
data_analyst.head()
```

Clean the column "Strong Word".

```python
data_analyst['Strong Words'][0]
```

```python
type(data_analyst['Strong Words'][0])
```

```python
strong = []
replacement = {"[":"","]":"","'":"",",":""}
for i in data_analyst['Strong Words']:
    try:
        strong.append(replace(i,replacement))
    except:
        strong.append('N/A')
data_analyst['Strong Words']=strong
```

```python
data_analyst.head()
```

Now I'm going to filter out the most related data analyst jobs. Because of the recommendation system of Seek.com.au, it will return some relevant jobs like Front End Developer and IT Analyst because in the job descriptions of these jobs also contains the keyword "Data Analyst". I will filter out the most relevant Data Analyst jobs for our analysis later on, which are the ones explicitly contains "Data Analyst" in their titles.

```python
data_analyst = data_analyst[data_analyst['Job Title'].str.contains('Data Analyst') | data_analyst['Job Title'].str.contains('Data Analytics') | data_analyst['Job Title'].str.contains('Statistical Analyst')]
```

```python
data_analyst['Category'] = 'Data Analyst'
```
